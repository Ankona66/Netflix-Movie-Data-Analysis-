```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
#Dataframe
df = pd.read_csv('mymoviedb.csv', lineterminator= '\n')
df.head() 
```

```python
#Summary of work TBD
# Release date to be casted in Date and time format
# overview, Original Language, Poster Url can be removed
# Genre column has a white space and that needs to be removed if required

#Changing the format of Release date to Date Time
df['Release_Date'] = pd.to_datetime(df['Release_Date'])
print(df['Release_Date'] )

#Just extracting the year
df['Release_Date'] = df['Release_Date'].dt.year 
```

```python
df.head()
```

```python
#Removing the columns
cols = ['Overview', 'Original_Language', 'Poster_Url']
df.drop(cols, axis=1, inplace = True)
df.head()
```

Categorizing the Vote_Average Column in 4 categories

```python
def categorize_col(df, col, labels):
    edges = [df[col].describe()['min'], 
    df[col].describe()['25%'],
    df[col].describe()['50%'],
    df[col].describe()['75%'],
    df[col].describe()['max']]
    df[col] = pd.cut(df[col], edges, labels = labels, duplicates = 'drop')
    return df
```

```python
labels = ['not_popular', 'below_avg', 'average', 'popular']
categorize_col(df, 'Vote_Average',labels)
df['Vote_Average'].unique()
```

```python
#To drop the null values
df.dropna(inplace = True)
df.isna().sum()
```

Removing spaces from genre and splitting genre into list

```python
df['Genre'] = df['Genre'].str.split(', ')
df = df.explode('Genre').reset_index(drop= True)
df.head()
#Casting Genre into category
df['Genre'] = df['Genre'].astype('category')
df['Genre'].dtypes
```
